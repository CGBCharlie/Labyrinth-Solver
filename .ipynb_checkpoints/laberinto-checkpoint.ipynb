```python
class Nodo:
    def __init__(self, valor):
        self.valor= valor
        self.padre= None
        self.hijos= []
        
    def buscarPadre(self, valor):
        if self.valor == valor:
            return self
        else:
            padre = None
            for hijo in self.hijos:
                padre = hijo.buscarPadre(valor)
                if padre: 
                    break
            return padre
        return None

class Arbol:
    def __init__(self):
        self.raiz = Nodo(None)
        
    def agregarHijo(self, valor, padre):
        nodo = Nodo(valor)
        padre.hijos.append(nodo)
        nodo.padre = padre
    
    def ancestros(self, explorador):
        resultado = [explorador.valor]
        while explorador != self.raiz:
            explorador = explorador.padre
            resultado += [explorador.valor]   
        return resultado 
    
    def buscarFondo(self, raiz, fondo):
        if raiz.hijos:
            for hijo in raiz.hijos:
                if hijo.valor == fondo.valor:
                    self.ancestros(hijo)
                else:
                    self.buscarFondo(hijo, fondo)
        else:
            return
```
